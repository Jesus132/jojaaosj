# BUSQUEDA DEL ALPINISTA Y RECOCIDO SIMULADO

En este documento nos encontramos con unos algoritmos
que buscara el camino mas optimo teniendo en cuenta
unas pautas ya que sabiendo las coordenadas de cada nodo
podemos calcular la distancia con la formula punto a punto
y teniendo la distancia podemos usar estos 2 algoritmos de
busqueda del camino mas optimo.

```python
#Alpinista
import math
import random as rd

def distancesFromCoords():
    f = open('berlin52.tsp')
    data = [line.replace("\n","").split(" ")[1:] for line in f.readlines()[6:58]]
    coords =  list(map(lambda x: [float(x[0]),float(x[1])], data))
    distances = []
    for i in range(len(coords)):
        row = []
        for j in range(len(coords)):
            row.append(math.sqrt((coords[i][0]-coords[j][0])**2 + (coords[i][1]-coords[j][1])**2))
        distances.append(row)
    return distances

def P(Rutaaux, Ruta):
  Rutaaux = Ruta.copy()
  r1= rd.sample(Rutaaux,1)
  r2 = rd.sample(Rutaaux,1)
  while r1[0] == 0 or r2[0] == 0:
     r1= rd.sample(Rutaaux,1)
     r2= rd.sample(Rutaaux,1)
  
  for i in range(len(Rutaaux)):
    if(Rutaaux[i] == r1[0]):
      x=i
    if(Rutaaux[i] == r2[0]):
      y=i
  
  aux=Rutaaux[x]
  Rutaaux[x] = Rutaaux[y]
  Rutaaux[y] = aux
  return Rutaaux

def Alpinista(Rutaaux, Ruta, interacciones, distances):
  for i in range(interacciones):
    Rutaaux=P(Rutaaux,Ruta)
    SumaRandm1 = 0.0
    SumaRandm2 = 0.0
    for i in range(len(distances)-1):
      SumaRandm1 += distances[Rutaaux[i]][Rutaaux[i+1]]
      SumaRandm2 += distances[Ruta[i]][Ruta[i+1]]
    if SumaRandm1 < SumaRandm2:
      Ruta = Rutaaux
  return Ruta, SumaRandm2

if __name__ == '__main__':
  Ruta = [i for i in range(52)] + [0]
  Rutaaux=[]
  Ruta_Optima, Suma_Optima=Alpinista(Rutaaux, Ruta, 10000, distancesFromCoords())
  print("Ruta optima:",Ruta_Optima)
  print("Solucion optima:",Suma_Optima)
```

```python
#Recocido Simulado
import math
import random as rd

def distancesFromCoords():
    f = open('berlin52.tsp')
    data = [line.replace("\n","").split(" ")[1:] for line in f.readlines()[6:58]]
    coords =  list(map(lambda x: [float(x[0]),float(x[1])], data))
    distances = []
    for i in range(len(coords)):
        row = []
        for j in range(len(coords)):
            row.append(math.sqrt((coords[i][0]-coords[j][0])**2 + (coords[i][1]-coords[j][1])**2))
        distances.append(row)
    return distances

def P(Rutaaux, Ruta):
  Rutaaux = Ruta.copy()
  r1= rd.sample(Rutaaux,1)
  r2 = rd.sample(Rutaaux,1)
  while r1[0] == 0 or r2[0] == 0:
     r1= rd.sample(Rutaaux,1)
     r2= rd.sample(Rutaaux,1)
  
  for i in range(len(Rutaaux)):
    if(Rutaaux[i] == r1[0]):
      x=i
    if(Rutaaux[i] == r2[0]):
      y=i
  
  aux=Rutaaux[x]
  Rutaaux[x] = Rutaaux[y]
  Rutaaux[y] = aux
  return Rutaaux


def Recocido(Rutaaux, Ruta, distancias, T):
  ite=0
  while(T >= 0.1):
    Rutaaux=P(Rutaaux, Ruta)
    ite+=1
    SumaRandm1=0.0
    SumaRandm2=0.0
    for i in range(len(distancias)-1):
      SumaRandm1 += distancias[Rutaaux[i]][Rutaaux[i+1]]
      SumaRandm2 += distancias[Ruta[i]][Ruta[i+1]]
    if (SumaRandm1 < SumaRandm2):
      Ruta=Rutaaux.copy()
      T *= 0.9999
    else:
      fact_enfriamiento=rd.uniform(0.8,0.99)
      if(fact_enfriamiento < math.exp(-( (SumaRandm1 - SumaRandm2) / T))):
          Ruta=Rutaaux.copy()
          T *= 0.9999
    print("T: ", T, "iteracion: ", ite)
  return Ruta, SumaRandm2


if __name__ == '__main__':
  Ruta = [i for i in range(52)] + [0]
  Rutaaux=[]
  Ruta_Optima, Suma_Optima=Recocido(Rutaaux, Ruta, distancesFromCoords(), 0.11)
  print("Ruta optima:",Ruta_Optima)
  print("Solucion optima:",Suma_Optima)
```
