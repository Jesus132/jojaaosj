```python
import math
import random as rd

T = [[0.0 for i in range(52)] for i in range(52)]
summ=[]
camm=[]

def distancesFromCoords():
    f = open('berlin52.tsp')
    data = [line.replace("\n","").split(" ")[1:] for line in f.readlines()[6:58]]
    coords =  list(map(lambda x: [float(x[0]),float(x[1])], data))
    distances = []
    N=[]
    for i in range(len(coords)):
        row = []
        aux=[]
        for j in range(len(coords)):
            row.append(math.sqrt((coords[i][0]-coords[j][0])**2 + (coords[i][1]-coords[j][1])**2))
        distances.append(row)
        #print(row[:len(row)])
        for i in row:#Recorremos la lista
          if(i != 0):#Python no divide entre 0
            aux.append(1/i)#Guardamos el valor en una lista aux
          else:
            aux.append(0.0)
        N.append(aux)
    return distances, N
#Funcion loca que se debe solo ignorar intente guardar las ciudades para ir eligiendo pero no era para las exploradoras
def funNodos(No, i):
  for x in No:
    if(i != x):
      return i
    else:
      return funNodos(No, i+1)

def Feromonas(Cam, Valor):
  for i in range(len(Cam)-1):
    T[Cam[i]][Cam[i+1]] += 1/Valor
  p=0.1#rd.random()
  for i in range(len(T)):
    for j in range(len(T)):
      T[i][j] *= (1-p)

#Funcion que reducira las feromonas al 90%
def HormigaExploradoras(Dist, Nod):
  for H in range(2000):
    Suma=0
    Camino = [funNodos(Nod, i+1)  for i in range(52 - len(Nod))]
    Camino = Nodo[-1:] + rd.sample(Camino, k = 52 - len(Nod)) + Nodo[-1:]#rd.sample(Camino[1:52], k=51)
    for i in range(len(Camino)-1):
      Suma += Dist[Camino[i]][Camino[i+1]]#Se guarda por el camino donde va la hormiga
    for i in range(len(Camino)-1):
      T[Camino[i]][Camino[i+1]] += 1/Suma
    #Feromonas(Camino, Suma)

def HormigaR(Dij, Nij):
  for hormigas in range(2000):
    Nodo=[0]#Lista que guarda las ciudadades que elige cada hormiga
    for i in range(1, 52):
      TN = [ i if i not in Nodo else 0 for i in range(52)]#Si i no esta en Nodo guarda i de otro modo se guarda 0 para el rango de 52
      for tn in TN:
        if(tn != 0):#Python no divide entre 0
          TN[TN.index(tn)] = T[Nodo[-1]][TN.index(tn)] * Nij[Nodo[-1]][TN.index(tn)]#Se guarda la probabilidad en la ciudad donde va el camino
      v=rd.random()
      cont=0
      for tn in TN:
        if(tn != 0):#Python no divide entre 0
          cont += tn/sum(TN)
          if(v < cont):
            Nodo.append(TN.index(tn))#Guardamos la ciudad donde entro
            break
    CAMINO = Nodo[0:len(Nodo)] + Nodo[:1]#Creamos el camino
    Suma=0
    for su in range(len(CAMINO) -1):
      Suma += Dij[CAMINO[su]][CAMINO[su+1]]#Sumamos el camino elegido por la hormiga
    summ.append(Suma)
    camm.append(CAMINO)
    Feromonas(CAMINO, Suma)#Reducimos las feromodas





if __name__ == '__main__':
  Nodo=[0]
  Dij,Nij=distancesFromCoords()#Sacamos las distancias de "berlin52" y N
  HormigaExploradoras(Dij, Nodo)#Mandamos Exploradoras para llenar la feromodas
  HormigaR(Dij, Nij)#Mandamos hormigas a que recorran por las feromonas
  print(camm[summ.index(min(summ))])#Imprimimos el Camino de la ciudad con menos valor
  print(min(summ))#Imprimimos el menor valor
```
