```python
import numpy as np
from math import exp

def MatrizConfusion(yaux):
	TN, FN, FP, TP = 0,0,0,0
	for i in range(len(yaux[0])):
		if yaux[1][i] == 1.0:
			if yaux[2][i] == 1.0:
				TP += 1
			else:
				FP += 1
		else:
			if yaux[2][i] == 1.0:
				FN += 1
			else:
				TN += 1
	print("Matriz: ")
	print("True negatives: ",TN)
	print("False positives: ",FP)
	print("False negatives: ",FN)
	print("True positives: ",TP)
	print("\n")
	#Accuracy
	accuracy = (TN + TP) / (TN + FP + FN + TP)
	#Recall
	recall = TP / (TP + FN)

	#Precision
	precision = TP / (TP+FP)

	#Specificity
	specificity = TN / (TN+FP)

	#F1
	F1score = ((2*precision*recall)/(precision + recall))

	print("Accuracy: ",accuracy)
	print("Recall: ",recall)
	print("Precision: ",precision)
	print("specificity: ",specificity)
	print("F1-Score: ",F1score)

def CrossEntropyError(y):
    E = 0
    for i in y[0]:
        if (i == 1):
            E += -np.log(i)
        else:
            E += -np.log(1 - i)
    print("Error: ", E)
    return y

def logi(xt):
    P = 1/(1 + np.exp(-xt))
    return P

def Eva(npx, npy, teta):
	s=[[],[],[]]
	for i in range(30):
		XT = np.dot(teta.T, npx[i])
		salida = logi(XT)
		if(salida > 0.5):
			s[0].append(salida)
			s[1].append(1)
			s[2].append(int(npy[i]))
			salida = 1
		else:
			s[0].append(salida)
			s[1].append(0)
			s[2].append(int(npy[i]))
			salida = 0
		print("Salida Real: " + str(int(npy[i])) + "\t Salida Evaluada: " + str(salida) + "\n")
	return s


def Teta(X, Y):
	TetaY=[]
	NPX=np.array(X)
	NPY=np.array(Y)
	TetaUna=np.dot(np.linalg.inv(np.dot(NPX.transpose(),NPX)),np.dot(NPX.transpose(),NPY))
	return TetaUna


def GradienteD(x,y,t,alfa,iterations):
     m=len(y)
     for it in range(iterations):
         temporal=np.copy(t)
         z=np.dot(x,t)
         h=logi(z)
         for j in range(t.shape[0]):
             suma=0
             for i in range(m):
                 suma=(h[i][0]-y[i])*x[i][j]*(1/m)+suma
             temporal[j]=t[j]-alfa*suma      
         t=np.copy(temporal)
     return t

if __name__ == '__main__':
	X, x=[], []#X para guardar los del 70% y x para los del 30%
	Y, y=[], []#Y para los del 70% y y para los del 30%
	aX = np.loadtxt('grades_data.txt' ,dtype=float, delimiter=',')#Guardamos los datos de un .txt en una np.array
	#np.random.shuffle(aX)#Desordenamos las listas de la matriz
	for i in range(len(aX)):#Recorremos la lista
		if(i < round(0.7*len(aX))):#los primeros 31 elementos lo guardara en X y Y
			X.append([1] + list(aX[i][:2]))
			Y.append(aX[i][-1])
		else:#los ultimos 13 lo guardara en x y y
			x.append([1] + list(aX[i][:2]))
			y.append(aX[i][-1])
	Te=Teta(X, Y)
	MatrizConfusion(CrossEntropyError(Eva(np.array(x), np.array(y), Te)))
	t=np.ones((NPX.shape[1],1))
	alfa=0.005
	T=GradienteD(NPX,NPY,t,alfa,2000000)
	print("-Gradiente Descendente-------------------------------------------------------------")
	MatrizConfusion(CrossEntropyError(Eva(np.array(x), np.array(y), T)))
```
